```python
# Counting genres
```

```python
movies_df["genres"].tail()
```

```python
movies_genres = {}
```

```python
for i in movies_df["genres"]:
    i = i.lstrip("['").rstrip("']").split("', '")
    for j in i:
        if j != "":
            movies_genres[j] = movies_genres.get(j, 0) + 1
```

```python
movies_genres
```

```python
movies_genres_df = pd.DataFrame(list(movies_genres.items()), columns = ["genre", "count"])
```

```python
movies_genres_df
```

```python
movies_genres_df.dtypes
```

```python
# Pie chart from movies_genres_df
```

```python
plt.pie(movies_genres_df["count"], labels = movies_genres_df["genre"])
plt.show()
```

```python
# Making a DF based on age_certification
```

```python
movies_df["age_certification"].unique()
```

```python
movies_age_certification = {}
```

```python
for i in movies_df["age_certification"]:
    if str(i) != "nan":
        movies_age_certification[i] = movies_age_certification.get(i, 0) + 1
```

```python
movies_age_certification
```

```python
movies_age_cert_df = pd.DataFrame(list(movies_age_certification.items()), columns = ["age_certification", "count"])
```

```python
movies_age_cert_df
```

```python
# Making a pie chart of age certification
```

```python
plt.pie(movies_age_cert_df["count"], labels = movies_age_cert_df["age_certification"])
plt.show()
```

```python
# Grouping by the highest rated movies by year
```

```python
movies_by_year_df = movies_df.groupby("release_year").agg({"imdb_score" : "max", "title" : "first"}).sort_values(by = "release_year").reset_index()
```

```python
movies_by_year_df
```

```python
movies_by_year_df.dropna(axis = 0, how = "any").reset_index()
```

```python
# Making a scatter plot for movies_by_year_df
```
